# Fundamental Data Type:
- **int**: Integer numbers, e.g., 1, 2, 3
- **float**: Floating-point numbers, e.g., 1.0, 2.5, 3.14
- **str**: Strings, e.g., "hello", "world"
- **bool**: Boolean values, e.g., True, False
- **list**: Ordered collection of items, e.g., [1, 2, 3], ["apple", "banana"]
- **tuple**: Immutable ordered collection of items, e.g., (1, 2, 3), ("apple", "banana")
- **dict**: Key-value pairs, e.g., {"name": "Alice", "age": 30}
- **set**: Unordered collection of unique items, e.g., {1, 2, 3}, {"apple", "banana"}

#Classes -> Custom Data Types
- **class**: A blueprint for creating objects, e.g., class Dog: pass
- **object**: An instance of a class, e.g., my_dog = Dog()
- **method**: A function defined within a class, e.g., def bark(self): pass
- **attribute**: A variable defined within a class, e.g., self.name = "Buddy"

#special data types
