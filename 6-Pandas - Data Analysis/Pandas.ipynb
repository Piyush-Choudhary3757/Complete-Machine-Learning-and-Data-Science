044-Manipulating Data 2-opendir.cloud


```python
car_sales
```

```python
# Column from series
series_column = pd.Series([5, 5, 5, 5, 5])
# New column called "Seats"
car_sales["Seats"] = series_column
car_sales
```

```python
car_sales.fillna({"Seats": 5}, inplace=True)
```

```python
car_sales
```

```python
# column from Python list
fule_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fule_economy
car_sales
```

```python
# fule use in life time
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"] / \
    100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
# create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
# put toyota car have 4 wheels and bmw have 3 wheels
car_sales["Number of wheels"] = car_sales["Make"].apply(
    lambda x: 4 if x == "toyota" else 3)
car_sales
```

```python
# put toyota and nissan car have 4 wheels and bmw have 3 wheels and honda have 2 wheels
car_sales["Number of wheels"] = car_sales["Make"].apply(
    lambda x: 4 if x == "toyota" or x == "nissan" else 3 if x == "bmw" else 2)
car_sales
```

```python
car_sales["passed road safety"] = True
car_sales.dtypes
```

```python
car_sales = car_sales.drop("Total fuel used (L)", axis=1, errors='ignore')
car_sales
```
